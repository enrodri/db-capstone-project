```python
# We import mysql.connector 
import mysql.connector as connector
from datetime import datetime, timedelta
```

```python
# We establish kwargs for connection
connection=connector.connect(
    database="little_lemon_schema",
    user="little_lemon",
    password="capstone",
    auth_plugin='mysql_native_password'
)
```

```python
# Create the cursor
cursor = connection.cursor()
cursor.execute("USE little_lemon_schema")
# Confirm the database in use
print("Database in use is:", connection.database)
```

```python
# This is a query that returns all the tables existing on the database
show_tables_query = "SHOW tables" 
cursor.execute(show_tables_query)
results = [row[0] for row in cursor.fetchall()]
print("Current tables in database: \n", results)
```

```python
# JOIN query for customers' details where Total Cost is higher than 60
greater_than_sixty = """
SELECT CONCAT(cd.FirstName, ' ', cd.LastName) AS 'Full Name', cd.Phone, cd.Email, o.BillAmount AS 'Total Cost'
FROM CustomerDetails cd
INNER JOIN Orders o
ON cd.CustomerID=o.CustomerID
WHERE o.BillAmount >= 60;
"""
cursor.execute(greater_than_sixty)
# Fetch all results that satisfy the query 
results = cursor.fetchall()

# Retrieve column names
cols = cursor.column_names

# Print column names and records from results using for loop
print("""Customers' details with Total Cost higher than $60:""")
print(cols)
for result in results:
    print(result)
```
